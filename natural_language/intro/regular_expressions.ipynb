```python
import nltk
import re
```

```python
nltk.download('cess_esp')
```

Regular Expresions

Flatten

```python
corpus=nltk.corpus.cess_esp.sents() #corpus is a large set of text
flatten=[w for l in corpus for w in l] #list comperhentions
```

re.search(p,s) 
this library has 2 parameters
p=search patro & s String

```python
arr=[w for w in flatten if re.search("ar$",w)]
print(arr)
#if we use "^" the regular expresion is required to state in the init
#if we use "$" the regular expresion is required to state in the end
#using "ar" regular expresion we try to have jus infinitive verbs
```

re.search in range

```python
arr=[w for w in flatten if re.search("^[ghi]",w)]
print(arr)
```

Closing
* repeat 0 or more times || required for true
+ repeat 1 or more times || required for true

```python
arr=[w for w in flatten if re.search('^(no)+',w)]
print(arr[:20])
```
